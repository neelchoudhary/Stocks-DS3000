```python
# Import for libraries we use
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as scipy
import matplotlib.pyplot as plt
```

```python
# Return the list of data points to delete
def delete_col_row(row_index_del, num_years, row, num_rows_in_sheet):
    li = []
    for num in range(0, num_rows_in_sheet):
        del_index = int((num*num_years) + row_index_del)
        if ((del_index+num) < row.size):
            li.append(del_index+num)
    return li
```

```python
# Read list of tickers from github
df_tickers = pd.read_csv("https://raw.githubusercontent.com/neelchoudhary/Stocks-DS3000/master/Stock%20Tickers.csv", index_col=None, header=0)
# Partial path of each csv in github folder
path = "https://raw.githubusercontent.com/neelchoudhary/Stocks-DS3000/master/stock_data/"
incomes_li = []
balances_li = []
flows_li = []
incomes_names = []
balances_names = []
flows_names = []
# Iterate through list of tickers from data frame df_tickers
for ticker in list(df_tickers["Tickers"]):
    # Full path of each csv file in github
    filename = path + ticker + ".csv"
    df = pd.read_csv(filename, index_col=None, header=0).T
    filename = filename.replace("https://raw.githubusercontent.com/neelchoudhary/Stocks-DS3000/master/stock_data/", "")
    # Income statement
    if "_1" in filename:
        # Calculate number of years given in this csv
        num_years = ((df.shape[1])/20) - 1 # Exclude TTM
        # If there is less than 15 years (+ TTM), we don't consider the stock. 
        if (num_years >= 16):
            s = df.iloc[0]
            list_to_drop = delete_col_row(0, num_years, s, 20)
            initial_count = 16
            # We want to disregard 2019 data
            if ("/2019" in s[1]):
                initial_count = 17
                list_to_drop = list_to_drop + delete_col_row(1, num_years, s, 20)
            # Drop all the data before 2004, so that we only have 15 years worth of data for each stock. 
            for num in range(initial_count, int(num_years) + 1):
                e = delete_col_row(num, num_years, s, 20)
                list_to_drop = list_to_drop + e;
            s = s.drop(s.index[list_to_drop])
            s = s.reset_index(drop=True)
            incomes_li.append(s)
            incomes_names.append(filename.replace("_1.csv", ""))
    # Balance sheet
    elif "_2" in filename:
        # Same as before.
        num_years = ((df.shape[1])/35)
        if (num_years >= 16):
            s = df.iloc[0]
            list_to_drop = []
            initial_count = 16
            if ("/2019" in s[0]):
                initial_count = 17
                list_to_drop = list_to_drop + delete_col_row(0, num_years - 1, s, 35)
            for num in range(initial_count - 1, int(num_years)):
                e = delete_col_row(num, num_years - 1, s, 35)
                list_to_drop = list_to_drop + e;
            s = s.drop(s.index[list_to_drop])
            s = s.reset_index(drop=True)
            balances_li.append(s)
            balances_names.append(filename.replace("_2.csv", ""))
    # Cash flows statement
    elif "_3" in filename:
        # Same as before. 
        num_years = ((df.shape[1])/30) - 1 # Exclude TTM
        if (num_years >= 16):
            s = df.iloc[0]
            list_to_drop = delete_col_row(0, num_years, s, 30)
            initial_count = 16
            if ("/2019" in s[1]):
                initial_count = 17
                list_to_drop = list_to_drop + delete_col_row(1, num_years, s, 30)
            for num in range(initial_count, int(num_years) + 1):
                e = delete_col_row(num, num_years, s, 30)
                list_to_drop = list_to_drop + e;
            s = s.drop(s.index[list_to_drop])
            s = s.reset_index(drop=True)
            flows_li.append(s)
            flows_names.append(filename.replace("_3.csv", ""))

# Create dataframes from scraped data
incomes = pd.DataFrame(incomes_li);
balances = pd.DataFrame(balances_li);
flows = pd.DataFrame(flows_li);

# Set the index to their respective ticker names.
incomes.index = incomes_names
balances.index = balances_names
flows.index = flows_names
# Sort the index tickers in alphabetical order
incomes = incomes.sort_index()
balances = balances.sort_index()
flows = flows.sort_index()
names = incomes.index
```

```python
# Outputs a series of values for a specific data point for a specific stock given it's row series.
# Also cleans the data by removing commas. 
def get_values(datapoint_index, row, num_years):
    start_index = int(num_years*datapoint_index)
    lis = []
    for num in range(int(start_index),int(15+start_index)):
            lis.append(row[num].replace(',',''))
    series = pd.Series(lis)
    return series

# Gets the specific row series from the data frame given the row number. 
def get_index_values(datapoint_index, row_num, df, row_qty):
    row = df.iloc[row_num]
    num_years = (pd.DataFrame(row).last_valid_index()+1)/row_qty
    return get_values(datapoint_index, row, num_years)

# These represent the data points we want to select, and returns a series row for the given stock index.
def make_row(li, index, df, row_qty):
    series = pd.Series()
    for l in li:
        series = series.append(get_index_values(l, index, df, row_qty),ignore_index=True)
    return series
    
# Return a data frame of all the data points of a specfic statement we want to select for each stock.
def make_df(li, df, row_qty):
    df_new = pd.DataFrame()
    i = 0
    # Use manual index, i, because given index is the stock ticker.
    for index, row in df.iterrows():
        s = make_row(li,i,df,row_qty)
        df_new = df_new.append(s,ignore_index=True)
        i += 1
    return df_new


# List of datapoint indices from yahoo finance that represent the features we want: 

# Total Revenue (1), Cost of Revenue (2), Operating Income (7), Interest Expense (8), Net Income (13)
interested_indexes_income_statement = [1, 2, 7, 8, 13]

# Total Cash (3), Net Receivables (4), Total Current Assets (7), Total Assets (16), Total Revenue (17),
# Total Current Liablities (23), Long Term Debt (24), Total Liablities (29), Total Stockholder's Equity (33)

interested_indexes_balance_statement = [3, 4, 7, 16, 17, 23, 24, 29, 33]
# Depreciation & Amortization (2)
interested_indexes_flow_statement = [2]

# Filtered statements with the datapoints we want:

incomes_filtered = make_df(interested_indexes_income_statement, incomes, 20)
incomes_filtered.index = names

balances_filtered = make_df(interested_indexes_balance_statement, balances, 35)
balances_filtered.index = names

flows_filtered = make_df(interested_indexes_flow_statement, flows, 30)
flows_filtered.index = names
```

We filter our data frames to include only the features that we want to target in our analysis. Next, we will convert these raw financial numbers into financial ratios which we can further analyze.  

```python
# Helper functions for converting to ratios

# Outputs a series of values representing 15 year's worth of data for a feature. 
# If the value is a dash (representing an empty value), then we consider it as 0. 
def get_values_filtered(datapoint_index, row, num_years):
    start_index = int(num_years*datapoint_index)
    lis = []
    for num in range(int(start_index),int(15+start_index)):
        if (row[num] == "-"):
            lis.append(0)
        elif (type(row[num]) == str):
            lis.append(int(row[num].replace(',','')))
        else:
            lis.append(row[num])
    series = pd.Series(lis)
    return series


# Gets the specific row series from the data frame given the row number. 
def get_index_values_filtered(datapoint_index, row_num, df, row_qty):
    row = df.iloc[row_num]
    num_years = (pd.DataFrame(row).last_valid_index()+1)/row_qty
    return get_values_filtered(datapoint_index, row, num_years)

# Get the financial average for a row (average between two years worth of data)
def get_average(row):
    avg_s = []
    prev = row[0]
    for curr in row[1:].iteritems():
        avg = (curr[1] + prev) / 2
        prev = curr[1]
        avg_s.append(avg)
    avg_s.append(row[row.size-1])
    return pd.Series(avg_s)
```

```python
# Calculate Financial Ratios
inc_df = incomes_filtered
bal_df = balances_filtered
flow_df = flows_filtered

def roe(stock_index):
    # Net income / Avg total SE
    total_se_s = pd.Series()
    total_se_s = total_se_s.append(get_index_values_filtered(8, stock_index, bal_df, 9),ignore_index=True)
    avg_total_se_s = get_average(total_se_s)
    net_income_s = pd.Series()
    net_income_s = net_income_s.append(get_index_values_filtered(4, stock_index, inc_df, 5),ignore_index=True)
    roe_s = pd.Series()
    roe_s = net_income_s / avg_total_se_s 
    return roe_s

def roa(stock_index):
    # Net income / Avg total assets
    net_income_s = pd.Series()
    net_income_s = net_income_s.append(get_index_values_filtered(4, stock_index, inc_df, 5),ignore_index=True)
    total_assets_s = pd.Series()
    total_assets_s = total_assets_s.append(get_index_values_filtered(3, stock_index, bal_df, 9),ignore_index=True)
    avg_total_assets_s = get_average(total_assets_s)
    roa_s = pd.Series()
    roa_s = net_income_s / avg_total_assets_s
    return roa_s

def financial_leverage(stock_index):
    # Avg total assets / Avg total SE
    total_assets_s = pd.Series()
    total_assets_s = total_assets_s.append(get_index_values_filtered(3, stock_index, bal_df, 9),ignore_index=True)
    avg_total_assets_s = get_average(total_assets_s)
    total_se_s = pd.Series()
    total_se_s = total_se_s.append(get_index_values_filtered(8, stock_index, bal_df, 9),ignore_index=True)
    avg_total_se_s = get_average(total_se_s)
    financial_leverage_s = pd.Series()
    financial_leverage_s = avg_total_assets_s / avg_total_se_s
    return financial_leverage_s

def gross_margin(stock_index):
    # (Total Revenue - Cost of Revenue) / Total Revenue
    revenue_s = pd.Series()
    revenue_s = revenue_s.append(get_index_values_filtered(0, stock_index, inc_df, 5),ignore_index=True)
    cogs_s = pd.Series()
    cogs_s = cogs_s.append(get_index_values_filtered(1, stock_index, inc_df, 5),ignore_index=True)
    gross_margin_s = pd.Series()
    gross_margin_s = (revenue_s - cogs_s) / revenue_s
    return gross_margin_s
    
def operating_margin(stock_index):
    # Operating Income / Total Revenue
    operating_income_s = pd.Series()
    operating_income_s = operating_income_s.append(get_index_values_filtered(2, stock_index, inc_df, 5),ignore_index=True)
    revenue_s = pd.Series()
    revenue_s = revenue_s.append(get_index_values_filtered(0, stock_index, inc_df, 5),ignore_index=True)
    operating_margin_S = pd.Series()
    operating_margin_S = operating_income_s / revenue_s
    return operating_margin_S

def current_ratio(stock_index):
    # Current Assets / Current Liablities
    total_curr_assets_s = pd.Series()
    total_curr_assets_s = total_curr_assets_s.append(get_index_values_filtered(2, stock_index, bal_df, 9),ignore_index=True)
    total_curr_liabilities_s = pd.Series()
    total_curr_liabilities_s = total_curr_liabilities_s.append(get_index_values_filtered(5, stock_index, bal_df, 9),ignore_index=True)
    current_ratio_s = pd.Series()
    current_ratio_s = total_curr_assets_s / total_curr_liabilities_s
    return current_ratio_s

def quick_ratio(stock_index):
    # (Cash + Net Receivable) / Current Liabilities
    cash_s = pd.Series()
    cash_s = cash_s.append(get_index_values_filtered(0, stock_index, bal_df, 9),ignore_index=True)
    net_receivable_s = pd.Series()
    net_receivable_s = net_receivable_s.append(get_index_values_filtered(1, stock_index, bal_df, 9),ignore_index=True)
    total_curr_liabilities_s = pd.Series()
    total_curr_liabilities_s = total_curr_liabilities_s.append(get_index_values_filtered(5, stock_index, bal_df, 9),ignore_index=True)
    quick_ratio_s = pd.Series()
    quick_ratio_s = (cash_s + net_receivable_s) / total_curr_liabilities_s
    return quick_ratio_s

def interest_coverage(stock_index):
    # (Operating income + depreciation) / Interest expense
    operating_income_s = pd.Series()
    operating_income_s = operating_income_s.append(get_index_values_filtered(2, stock_index, inc_df, 5),ignore_index=True)
    dep_s = pd.Series()
    dep_s = dep_s.append(get_index_values_filtered(0, stock_index, flow_df, 1),ignore_index=True)
    int_exp_s = pd.Series()
    int_exp_s = int_exp_s.append(get_index_values_filtered(3, stock_index, inc_df, 5),ignore_index=True)
    interest_coverage_s = pd.Series()
    interest_coverage_s = (operating_income_s + dep_s) / int_exp_s
    return interest_coverage_s

def debt_to_equity(stock_index):
    # Total liablities / Total SE
    total_liablities_s = pd.Series()
    total_liablities_s = total_liablities_s.append(get_index_values_filtered(7, stock_index, bal_df, 9),ignore_index=True)
    total_se_s = pd.Series()
    total_se_s = total_se_s.append(get_index_values_filtered(8, stock_index, bal_df, 9),ignore_index=True)
    debt_to_equity_s = pd.Series()
    debt_to_equity_s = total_liablities_s / total_se_s
    return debt_to_equity_s
    
def lt_debt_to_equity(stock_index):
    # Total lt debt / Total SE
    total_lt_debt_s = pd.Series()
    total_lt_debt_s = total_lt_debt_s.append(get_index_values_filtered(6, stock_index, bal_df, 9),ignore_index=True)
    total_se_s = pd.Series()
    total_se_s = total_se_s.append(get_index_values_filtered(8, stock_index, bal_df, 9),ignore_index=True)
    lt_debt_to_equity_s = pd.Series()
    lt_debt_to_equity_s = total_lt_debt_s / total_se_s
    return lt_debt_to_equity_s
    
def profit_margin(stock_index):
    # Net income / Net Sales
    net_income_s = pd.Series()
    net_income_s = net_income_s.append(get_index_values_filtered(4, stock_index, inc_df, 5),ignore_index=True)
    revenue_s = pd.Series()
    revenue_s = revenue_s.append(get_index_values_filtered(0, stock_index, inc_df, 5),ignore_index=True)
    profit_margin_s = pd.Series()
    profit_margin_s = net_income_s / revenue_s
    return profit_margin_s
```

```python
# Calculate all ratios for a given stock index and returns a series.
def make_row_ratios(index):
    series = pd.Series()
    ratios = [roe, roa, financial_leverage, gross_margin, operating_margin, current_ratio, quick_ratio, debt_to_equity, lt_debt_to_equity, profit_margin]
    for ratio in ratios:
        series = series.append(ratio(index),ignore_index=True)
    return series
  
# Returns a data frame with all the ratios for each stock.
def all_ratios(df):
    df_ratios = pd.DataFrame()
    i = 0
    for index, row in df.iterrows():
        s = make_row_ratios(i)
        df_ratios = df_ratios.append(s,ignore_index=True)
        i += 1
    return df_ratios

# Return a list of column headers
def ratio_column_names():
    ratio_names = ["ROE", "ROA", "Financial Leverage", "Gross Margin", "Operating Margin", "Current Ratio", "Quick Ratio", "Debt to Equity", "Long Term Debt to Equity", "Profit Margin"]
    ratios_names_years = []  
    for name in ratio_names:
        for year in range(2018, 2004-1, -1):
            ratios_names_years.append(name + " - " + str(year))
    return ratios_names_years

ratios_df = all_ratios(incomes)
ratios_df.index = names
ratios_df.columns = ratio_column_names()
```

We now have a data frame of financial ratios for each stock for 15 years. Our next step is to perform a linear regression for each ratio over 15 years -- for each stock. We take the slope of the best-fit line and use that as a metric to indicate trend over time for each ratio for each stock. We refer to this slope as the "score". In this way, we have encapsulated 15 years worth of ratio data into one ratio metric (features).

```python
# Linear Regression for Ratio Score: 

# Performs linear regression on the given ratio (ratio_func) and stock index, and return the score (slope). 
def ratio_score(ratio_func, stock_index):
    # We pass our ratio function as a parameter, and compute it here to get s. 
    s = ratio_func(*stock_index)
    # Reverse the series so we go from 2004 to 2018. 
    s = s[::-1]
    ratio_trend = s.values
    years = [i for i in range(2004, 2019)]
    # Perform linear regression for the ratio values from 2004 to 2018. 
    slope, intercept, r_value, p_value, std_err = scipy.linregress(years, ratio_trend)
    return slope

# Find all the ratio scores for the given stock index, also check for NaN values and replace with 0. 
def total_score_stock(stock_index):
    s = pd.Series()
    score_list = []
    ratios = [roe, roa, financial_leverage, gross_margin, operating_margin, current_ratio, quick_ratio, debt_to_equity, lt_debt_to_equity, profit_margin]
    for ratio in ratios:
        score = ratio_score(ratio, [stock_index])
        # Check for NaN values, if NaN replace with 0. 
        if (score == score):
            score_list.append(score)
        else:
            score_list.append(0)
    return pd.Series(score_list)

# Return a data frame of scores for all the stocks. 
def all_scores(df):
    df_scores = pd.DataFrame()
    i = 0
    for index, row in df.iterrows():
        score = total_score_stock(i)
        df_scores = df_scores.append(score,ignore_index=True)
        i += 1
    return df_scores

# Return a list of column headers
def score_column_names():
    ratio_names = ["ROE", "ROA", "Financial Leverage", "Gross Margin", "Operating Margin", "Current Ratio", "Quick Ratio", "Debt to Equity", "Long Term Debt to Equity", "Profit Margin"]
    col_names = []  
    for name in ratio_names:
        col_names.append(name + "_score")
    return col_names

df_ratio_score = all_scores(incomes)
df_ratio_score.index = names
df_ratio_score.columns = score_column_names()
```

We now retrieve stock price history data from the github. We then perform the same linear regression approach on the stock price for each stock for 15 years from 2004 to 2018. This gives the stock price score (target). We then will append the stock price score column into our ratio score data frame to give us our final data frame. 

```python
# Retrieve stock price history from github into a data frame. 
stock_price_history_df = pd.read_csv("https://raw.githubusercontent.com/neelchoudhary/Stocks-DS3000/master/Stock%20Price%20History.csv", index_col="Ticker", header=0)
stock_price_history_df
```

```python
# Linear Regression for Price Score: 

# Performs linear regression on the given stock price dataframe and stock index, and return the price score (slope). 
def stock_price_score(stock_price_df, stock_index):
    s = stock_price_df.iloc[stock_index]
    # Reverse the series so we go from 2004 to 2018. 
    s = s[::-1]
    price_trend = s.values
    years = [i for i in range(2004, 2019)]
    # Perform linear regression for the stock price values from 2004 to 2018. 
    slope, intercept, r_value, p_value, std_err = scipy.linregress(years, price_trend)
    return slope

# Returns a data frame with all the stock price scores. 
def stocks_prices_scores(df):
    df_price_scores = pd.DataFrame()
    i = 0
    for index, row in df.iterrows():
        price_score = stock_price_score(df, i)
        s = pd.Series([price_score])
        df_price_scores = df_price_scores.append(s,ignore_index=True)
        i += 1
    return df_price_scores

df_price_scores = stocks_prices_scores(stock_price_history_df)
df_price_scores.index = names
df_price_scores.columns=["Price_score"]
```

## Model Construction

### Imports needed for Hypothesis Testing and Machine Learning 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

### Hypothesis Testing - Pearson's Correlation Tests

We want to conduct a correlation test to see if our hypothesis has merit, and if the correlations noticed in our data exploration are significant.

```python
print("Hypothesis Testing Results\n")

x1 = df_final["Gross Margin_score"].array
y1 = df_final["Long Term Debt to Equity_score"].array
res1 = scipy.pearsonr(x1, y1)
print("Pearson's R for Gross Margin and LT Debt to Equity: " + str(res1[0]))
print("P-value for Gross Margin and LT Debt to Equity: " + str(res1[1]) + "\n")

x2 = df_final["ROE_score"].array
y2 = df_final["Financial Leverage_score"].array
res2 = scipy.pearsonr(x2, y2)
print("Pearson's R for ROE and Financial Leverage: " + str(res2[0]))
print("P-value for ROE and Financial Leverage: " + str(res2[1]) + "\n")

x3 = df_final["Profit Margin_score"].array
y3 = df_final["Financial Leverage_score"].array
res3 = scipy.pearsonr(x3, y3)
print("Pearson's R for Profit Margin and Financial Leverage: " + str(res3[0]))
print("P-value for Profit Margin and Financial Leverage:: " + str(res3[1]) + "\n")

x4 = df_final["ROA_score"].array
y4 = df_final["Price_score"].array
res4 = scipy.pearsonr(x4, y4)
print("Pearson's R for ROA and Stock Price: " + str(res4[0]))
print("P-value for ROA and Stock Price: " + str(res4[1]))
```

The p-values for each of these tests are 6.761616695054517e-14, 3.9561530278774995e-49, 0.012741157616546123, and 0.1345520163235101.

The first three p-values are statistically significant since they are below 0.05, but the last one is not. We will explain the meaning of this in the discussion.

### Normalizing Data

```python
features = df_final.drop("Price_score", axis = 1)
target = df_final["Price_score"]

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=30000)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, random_state=30000)
```

We split our data into training, validation, and testing sets. 

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)

X_train_scaled = pd.DataFrame(X_train_scaled)
X_test_scaled = pd.DataFrame(X_test_scaled)
X_val_scaled = pd.DataFrame(X_val_scaled)
```

We choose to use the standard scaler so that the mean of the scores is 0 and the standard deviation of the scores is 1. This normalization is done in preparation for the machine learning. 

```python
ridge_model_init = Ridge().fit(X=X_train_scaled, y=y_train)
knn_model_init = KNeighborsRegressor().fit(X=X_train_scaled, y=y_train)
svr_model_init = SVR(gamma="scale").fit(X=X_train_scaled, y=y_train)
```

We applied three machine learning algorithms and trained them on our scaled data. 

## Model Evaluation

```python
print("Initial R-squared value for validation set: ", r2_score(y_val, ridge_model_init.predict(X_val_scaled)))
print("Initial R-squared value for validation set: ", r2_score(y_val, knn_model_init.predict(X_val_scaled)))
print("Initial R-squared value for validation set: ", r2_score(y_val, svr_model_init.predict(X_val_scaled)))
```

These were our results for our default machine learning algorithms -- Ridge, KNN, and SVR. 
We will now aim to optimize and tune the hyperparameters for our algorithms. 

## Model Optimization 

```python
# Ridge tuning optimization that returns the best alpha value. 
def ridge_tuning(start, end, jump):
    best_alpha = 0
    best_r2 = 0
    gap = -100000
    for a in np.arange(start, end, jump):
        m = Ridge(alpha=a).fit(X=X_train_scaled, y=y_train)
        r2score = r2_score(y_val, m.predict(X_val_scaled))
        if (r2score > gap):
            gap = r2score
            best_alpha = a
            best_r2 = r2score
    return best_alpha
    
best_alpha = ridge_tuning(0,10001,100)
print("Best alpha value: " + str(best_alpha))

ridge_model = Ridge(alpha=best_alpha).fit(X=X_train_scaled, y=y_train)

print("Tuned R-squared value for training set: ", r2_score(y_train, ridge_model.predict(X_train_scaled)))
print("Tuned R-squared value for validation set: ", r2_score(y_val, ridge_model.predict(X_val_scaled)))
```

```python
# KNN tuning optimization that returns the best n value. 
def knn_tuning(start, end, jump):
    best_n = 0
    best_r2 = 0
    gap = -10000
    for a in range(start, end, jump):
        m = KNeighborsRegressor(n_neighbors = a).fit(X=X_train_scaled, y=y_train)
        r2score = r2_score(y_val, m.predict(X_val_scaled))
        if (r2score > gap):
            gap = r2score
            best_n = a
            best_r2 = r2score
    return best_n


best_n = knn_tuning(2,100,1)
print("Best n value: " + str(best_n))

knn_model = KNeighborsRegressor(n_neighbors = best_n).fit(X=X_train_scaled, y=y_train)

print("Tuned R-squared value for training set: ", r2_score(y_train, knn_model.predict(X_train_scaled)))
print("Tuned R-squared value for validation set: ", r2_score(y_val, knn_model.predict(X_val_scaled)))
```

```python
# SVR tuning optimization that returns the best c value. 
def svr_tuning(start, end, jump):
    best_c = 0
    best_r2 = 0
    gap = -10000
    for a in range(start, end, jump):
        m = SVR(gamma="scale", C=a).fit(X=X_train_scaled, y=y_train)
        r2score = r2_score(y_val, m.predict(X_val_scaled))
        if (r2score > gap):
            gap = r2score
            best_c = a
            best_r2 = r2score
    return best_c

best_c = svr_tuning(8500,9000,1)
print("Best c value: " + str(best_c))
svr_model = SVR(gamma="scale", C=best_c).fit(X=X_train_scaled, y=y_train)

print("Tuned R-squared value for training set: ", r2_score(y_train, svr_model.predict(X_train_scaled)))
print("Tuned R-squared value for validation set: ", r2_score(y_val, svr_model.predict(X_val_scaled)))
```

We made a function for each of the three machine learning algorithms to tune the hyperparameters. Our desired goal was to push R-squared values for the validation set up. These were made with the logic behind GridSearchCV in mind (which tests different parameters and evaluates performance over k-folds). We ran this function a few times to narrow down the best hyperparameter. This gave us the best hyperparameters for each model. 

## Model Testing

These are our results for running the tuned machine learning models on our test set.

## 4. DISCUSSION

#### Summary 

Through exploring our data with visualizations we determined that we wanted to test correlations on different financial ratios to see which ones are related to each other. To do this, we created a heat map that showed the correlations between all the ratios. It was found that the following ratios are correlated: gross margin and long term debt to equity, return on equity and financial leverage, profit margin, and financial leverage. The previously mentioned correlations have all been proven to be statistically significant through a Pearson R correlation test. 

After conducting hypothesis testing we normalized our data based on the standard scalar method. This method arranged our data so that the mean was 0 and the standard deviation was 1. After normalizing our data, we split our data twice, to create a test, train, and validation set. After the split, we fit all of our data onto the chosen machine learning algorithms, tune them, and evaluate the results using the testing data.


#### Interpretation of Findings 

We compared three regression models: Ridge Linear Regression, KNN Regression, and SVR. 

We tuned the hyperparameters of each model to maximize the performance on our validation data set, which was to increase the R-squared value. For context, the R-squared value given from our machine learning models explain to what extent the variance of the changes in our stock fundamentals over time explains the variance of the stock price change over time.

When tuning, the SVR yielded the highest result, with R-squared close to 62 percent. This meant that approximately more than half of the observation variation can be explained by the model's inputs of stock fundamental change over time. However, once we reached Step 3.6 and evaluated the performance of these tuned models on our test data set, which acted as unseen data for the model, the R-squared scores for all three models dropped down close to 0. 

These R-squared values indicate that each model explains none of the variability of the response data around its mean.
In other words, none of the algorithms should be used for the predictive model since these algorithms cannot accurately predict changes in the target variable given the changes in the feature variables.

However, we were able to find and confirm, through Pearson’s r correlation hypothesis testing, that there is a statistically significant correlation between Gross Margin and Long Term Debt to Equity, between ROE and Financial Leverage, and between Profit Margin and Financial Leverage. This means for each pair listed, knowing how one changes over time can be used to guess how the other will change over time. There was a stastically significant correlation between ROA and Price, which means that knowing how one changes over time doesn't necessarily indicate how the other will.

 


#### Future Work

In the future, we should be more careful about our hypothesis and goals. While we did expect that changes in a stock’s fundamentals over time could be used to accurately predict changes in stock price over time, if this logic was accurate, it would have already been done by now by quantitative analysts and been widely used. Instead of focusing on the long term (15 years), a future project could focus on the short-term. This is because stock prices are very volatile in the short term, changing as people’s feelings about companies change in small periods of time. Perhaps a new study could conduct sentiment analysis through Twitter about how people feel regarding certain stocks, and analyzing short-term changes of that stock (up or down) to see if matched sentiments (positive and negative).
